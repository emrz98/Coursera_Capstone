```python
results = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
results
```

```python
df=results[0]
df.head()
```

```python
ndf=df.replace("Not assigned",np.nan)
ndf.head()
```

```python
ndf=ndf.dropna(subset=['Borough'],axis=0)
ndf=ndf.reset_index()
df_sin_drop=ndf.drop('index',axis=1)
df_sin_drop.head()
```

```python
#ndf.duplicated(ndf.columns[~ndf.columns.isin(['Postcode'])])
df_con_drop=df_sin_drop.drop_duplicates('Postal code')
df_con_drop=df_con_drop.reset_index()
df_con_drop=df_con_drop.drop('index',axis=1)
df_con_drop.head()
```

```python

for n in range(0,len(df_con_drop)):#len(df_sin_drop)):
    for i in range(0,len(df_sin_drop)):
        if(df_con_drop.loc[n,'Postal code']== df_sin_drop.loc[i,'Postal code'] and df_con_drop.loc[n,'Neighborhood']!=df_sin_drop.loc[i,'Neighborhood']):
            df_con_drop.loc[n,'Neighborhood']=df_con_drop.loc[n,'Neighborhood']+","+df_sin_drop.loc[i,'Neighborhood']
         
```
